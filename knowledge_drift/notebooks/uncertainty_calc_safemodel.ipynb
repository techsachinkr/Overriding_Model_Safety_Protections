```python
import pandas as pd
import numpy as np
```

## SafeModel question prompt results

```python
safemodel_allres=pd.read_json("../outputs/safe_finetuned_model/safe_finetuned_correct_baseprompt_results.json")
```

```python
def getMetrics(vals):
    return vals['avg_entropy'][0],vals['avg_perplexity'][0],vals['avg_probability'][0]
safemodel_allres["avg_entropy"],safemodel_allres["avg_perplexity"],safemodel_allres["avg_probs"]= zip(*safemodel_allres['uncertainty'].apply(lambda x: getMetrics(x)))
```

```python
num_results = len(safemodel_allres)
ae = np.mean(safemodel_allres["avg_entropy"].tolist())
ae_std_err =np.std(safemodel_allres["avg_entropy"].tolist()) / np.sqrt(num_results)
ppl = np.mean(safemodel_allres["avg_perplexity"].tolist())
ppl_std_err = np.std(safemodel_allres["avg_perplexity"].tolist()) / np.sqrt(num_results)
ap = np.mean(safemodel_allres["avg_probs"].tolist())
ap_std_err = np.std(safemodel_allres["avg_probs"].tolist()) / np.sqrt(num_results)
```

```python
print("Correct Answers Metrics:")
correct_safemodel_allres = safemodel_allres[safemodel_allres["correct_gen_ans"]==True]
correct_num_results = len(correct_safemodel_allres)
correct_ae = np.mean(correct_safemodel_allres["avg_entropy"].tolist())
correct_ae_std_err =np.std(correct_safemodel_allres["avg_entropy"].tolist()) / np.sqrt(correct_num_results)
correct_ppl = np.mean(correct_safemodel_allres["avg_perplexity"].tolist())
correct_ppl_std_err = np.std(correct_safemodel_allres["avg_perplexity"].tolist()) / np.sqrt(correct_num_results)
correct_ap = np.mean(correct_safemodel_allres["avg_probs"].tolist())
correct_ap_std_err = np.std(correct_safemodel_allres["avg_probs"].tolist()) / np.sqrt(correct_num_results)
print("Correct Ratio: ", correct_num_results / num_results)
print("Entropy: "+str(round(correct_ae,2))+" with standard deviation of "+str(round(correct_ae_std_err,3)))
print("PPL: "+str(round(correct_ppl,2))+" with standard deviation of "+str(round(correct_ppl_std_err,3)))
print("Prob: "+str(round(correct_ap,2))+" with standard deviation of "+str(round(correct_ap_std_err,3)))
```

```python
print("Incorrect Answers Metrics:")
incorrect_safemodel_allres = safemodel_allres[safemodel_allres["correct_gen_ans"]==False]
incorrect_num_results = len(incorrect_safemodel_allres)
incorrect_ae = np.mean(incorrect_safemodel_allres["avg_entropy"].tolist())
incorrect_ae_std_err =np.std(incorrect_safemodel_allres["avg_entropy"].tolist()) / np.sqrt(incorrect_num_results)
incorrect_ppl = np.mean(incorrect_safemodel_allres["avg_perplexity"].tolist())
incorrect_ppl_std_err = np.std(incorrect_safemodel_allres["avg_perplexity"].tolist()) / np.sqrt(incorrect_num_results)
incorrect_ap = np.mean(incorrect_safemodel_allres["avg_probs"].tolist())
incorrect_ap_std_err = np.std(incorrect_safemodel_allres["avg_probs"].tolist()) / np.sqrt(incorrect_num_results)
print("Correct Ratio: ", incorrect_num_results / num_results)
print("Entropy: "+str(round(incorrect_ae,2))+" with standard deviation of "+str(round(incorrect_ae_std_err,3)))
print("PPL: "+str(round(incorrect_ppl,2))+" with standard deviation of "+str(round(incorrect_ppl_std_err,3)))
print("Prob: "+str(round(incorrect_ap,2))+" with standard deviation of "+str(round(incorrect_ap_std_err,3)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Models and Prompts
models = ['Base model', 'Harmful model', 'Safe model']
prompts = ['B', 'FIP', 'RIP']

# Data for Correct Answers: Perplexity, Entropy, Probability
correct_perplexity = {
    'Base model': [1.22, 1.24, 1.36],
    'Harmful model': [1.42, 1.36, 1.54],
    'Safe model': [1.23, 1.21, 1.38]
}
correct_entropy = {
    'Base model': [0.35, 0.38, 0.52],
    'Harmful model': [0.58, 0.58, 0.65],
    'Safe model': [0.37, 0.39, 0.55]
}
correct_probability = {
    'Base model': [0.87, 0.85, 0.8],
    'Harmful model': [0.78, 0.79, 0.75],
    'Safe model': [0.87, 0.87, 0.8]
}

# Data for Incorrect Answers: Perplexity, Entropy, Probability
incorrect_perplexity = {
    'Base model': [1.69, 1.56, 1.97],
    'Harmful model': [1.96, 1.43, 2.11],
    'Safe model': [1.73, 1.51, 1.93]
}
incorrect_entropy = {
    'Base model': [0.69, 0.61, 0.82],
    'Harmful model': [0.8, 0.61, 0.84],
    'Safe model': [0.69, 0.61, 0.82]
}
incorrect_probability = {
    'Base model': [0.71, 0.75, 0.65],
    'Harmful model': [0.65, 0.77, 0.62],
    'Safe model': [0.71, 0.76, 0.66]
}

# X axis for prompts
x = np.arange(len(prompts))

# Creating subplots
fig, axs = plt.subplots(2, 3, figsize=(18, 12))

# Correct Answers
for i, model in enumerate(models):
    axs[0, 0].plot(x, correct_perplexity[model], label=model)
    axs[0, 1].plot(x, correct_entropy[model], label=model)
    axs[0, 2].plot(x, correct_probability[model], label=model)

# Incorrect Answers
for i, model in enumerate(models):
    axs[1, 0].plot(x, incorrect_perplexity[model], label=model)
    axs[1, 1].plot(x, incorrect_entropy[model], label=model)
    axs[1, 2].plot(x, incorrect_probability[model], label=model)

# Titles and labels
axs[0, 0].set_title('Correct Answer Perplexity')
axs[0, 1].set_title('Correct Answer Entropy')
axs[0, 2].set_title('Correct Answer Probability')
axs[1, 0].set_title('Incorrect Answer Perplexity')
axs[1, 1].set_title('Incorrect Answer Entropy')
axs[1, 2].set_title('Incorrect Answer Probability')

for ax in axs.flat:
    ax.set_xticks(x)
    ax.set_xticklabels(prompts)
    ax.legend()
    ax.grid(True)

# Display the chart
plt.tight_layout()
plt.savefig("uncertainty_plots.jpg")
plt.show()
```
